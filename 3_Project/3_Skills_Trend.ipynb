## How are in-demand skills trending for Data Analysts?

Methodology

- Aggregate skill counts monthly 
- Re-analyze based on percentage of total jobs   
- Plot monthly skill demand


```python
import ast
import pandas as pd
import seaborn as sns 
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x if isinstance(x, list) else None)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_mo'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_mo', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()
```

```python
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
df_DA_US_pivot.plot(kind='line', legend=False)


plt.show()
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_mo').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

df_DA_US_percent
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_mo'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_mo')

df_DA_US_percent

```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

plt.figure(figsize=(16, 9))

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US',  fontsize=17)
plt.ylabel('Likelihood in Job Postings', fontsize=16)
plt.xlabel('2025', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.legend(loc='upper right', fontsize='large')

from matplotlib.ticker import PercentFormatter
ax= plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


plt.show()
```
