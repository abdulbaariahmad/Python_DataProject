```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
median_yearly = df.salary_year_avg.median()
median_hourly = df.salary_hour_avg.median()
```

```python
# median_hourly = df.loc[:15, 'salary_hour_avg'].fillna(median_yearly)
# median_hourly = df.loc[:15, 'salary_hour_avg'].fillna(median_hourly)
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_yearly)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hourly)
```

```python
df_filled.loc[:10, 'salary_rate':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates()

print('Length of original df:       ', len(df_filled))
print("Length of drop duplicates df:", len(df_unique))
print("Number of removed raws:      ", len(df_filled)-len(df_unique))
```
